```python
# For visited nodes
visited = []

# For storing nodes to be explored
queue = []

# Breadth First Search
def bfs(visited, graph, node):
    
    
    path = {}
    
    
    
    # Add the first node to list of visited nodes
    visited.append(node)
    # Add the first node to the queue
    queue.append(node)
    
    # While the queue is not empty
    while queue:
        # Remove from the front of the queue
        m = queue.pop(0)
        print(m, end = "\n")
        
        # Looks at neighbors of current node
        for neighbour in graph[m]:
            # If the neighbor has not been visited
            if neighbour not in visited:
                
                
                path[neighbour] = m
                
                
                
                # Visit the neighbour
                visited.append(neighbour)
                # Add the neighbour to the queue to see if it has any neighbors
                queue.append(neighbour)
                
                
    print("Total Cost: ", len(visited)-1)
                
    # construct the path from the goal to the start
    finalPath = []
    
    # While we can still go backwards
    while m in path:
        #print("currNode: ", currNode)
        # Add the node
        finalPath.append(m)
        # Get the key that points to the node, this is the node's parent
        m = path[m]
        
    # Add the start state 
    finalPath.append(node)
    # Reverse the order
    finalPath.reverse()
    
    
    epicCounter = -1
    
    print("\n")
        
        
    for _ in finalPath:
        epicCounter = epicCounter + 1
        print(_)
        
    print("Best Solution Cost: ", epicCounter)

    
                
```

```python
# For visited nodes
visited = []

# Depth First Search
def dfs(visited, graph, node, path):
    
    
    
    
    # If node has not been visited
    if node not in visited:
        print(node, end = "\n")
        # Add node to visited
        visited.append(node)
        
        
        
        # Looks at neighbours of node
        for neighbour in graph[node]:
            
            
            if neighbour not in visited:
                path[neighbour] = node
            
            
            
            # Runs dfs again on neighbours
            dfs(visited, graph, neighbour, path)
            
    #return len(visited)
    #print(path)
    return path
            
    
```

```python
def getSuccessors(currentState):
    # Stores all possible successors (up to 4 possible for the 4 directions)
    successors = []
    # Stores index of number 0 (empty tile)
    emptyTileIndex = currentState.index(0)
    
    # Move empty tile up
    if emptyTileIndex != 0 and emptyTileIndex != 1 and emptyTileIndex != 2:
        # Create a copy of the list
        newState = list(currentState)
        # Swap the values
        temp = newState[emptyTileIndex]
        newState[emptyTileIndex] = newState[emptyTileIndex - 3]
        newState[emptyTileIndex - 3] = temp
        successors.append(newState)
        
         # Move empty tile down
    if emptyTileIndex != 6 and emptyTileIndex != 7 and emptyTileIndex != 8:
        # Create a copy of the list
        newState = list(currentState)
        # Swap the values
        temp = newState[emptyTileIndex]
        newState[emptyTileIndex] = newState[emptyTileIndex + 3]
        newState[emptyTileIndex + 3] = temp
        successors.append(newState)
    
    # Move empty tile left
    if emptyTileIndex != 0 and emptyTileIndex != 3 and emptyTileIndex != 6:
        # Create a copy of the list
        newState = list(currentState)
        # Swap the values
        temp = newState[emptyTileIndex]
        newState[emptyTileIndex] = newState[emptyTileIndex - 1]
        newState[emptyTileIndex - 1] = temp
        successors.append(newState)
    
    # Move empty tile right
    if emptyTileIndex != 2 and emptyTileIndex != 5 and emptyTileIndex !=  8:
        # Create a copy of the list
        newState = list(currentState)
        # Swap the values
        temp = newState[emptyTileIndex]
        newState[emptyTileIndex] = newState[emptyTileIndex + 1]
        newState[emptyTileIndex + 1] = temp
        successors.append(newState)
    
   
    
    #print(successors)
    #print("\n")
    return successors
```
